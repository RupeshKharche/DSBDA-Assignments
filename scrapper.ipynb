```python
from bs4 import BeautifulSoup
import requests
```

```python
# Define the url of the page
url = "https://www.amazon.in/Pixel-Charcoal-6GB-128GB-Stoarge/product-reviews/B0D2KYTZZS/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
# Getting the page and parsing it with html parser
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
# Extracting ratings and reviews from the page

# Extracting the whole list of reviews
reviews_list = soup.find('div', id='cm_cr-review_list')
```

```python
# Extracting divs containing actual reviews from the list
# review_containers = reviews_list.find("div", id=re.compile("customer_review-*"))
# review_containers
```

```python
# Extracting usernames from the review divs
users = []

user_spans = reviews_list.find_all("span", {"class": "a-profile-name"})

for span in user_spans:
	users.append(span.get_text())
```

```python
users
```

```python
# Extracting reviews from the review divs
reviews = []

review_spans = reviews_list.find_all("span", {"class": "review-text-content"})
```

```python
reviews.extend([span.get_text().strip("\n\t\r ") for span in review_spans])
```

```python
reviews
```

```python
# Extracting ratings from the review divs
ratings = []

rating_spans = reviews_list.find_all("span", {"class": "a-icon-alt"})
```

```python
for rating_span in rating_spans:
	ratings.append(rating_span.get_text().strip())
```

```python
ratings
```

```python
# Creating a dataframe from the data
import pandas as pd

df = pd.DataFrame({"Customer Name": users, "Review": reviews, "Rating": ratings})

df
```
